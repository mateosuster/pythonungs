# Gráficos

Un uso común para los portátiles es la visualización de datos mediante gráficos. La colaboración facilita esto con varias herramientas de gráficos disponibles como importaciones de Python.


## Matplotlib

Para graficar nuestros resultados utilizaremos la función `plot` de la librería `matplotlib`.

[Matplotlib](http://matplotlib.org/) es el paquete de gráficos más común; consulte la documentación de su función más utilizada [(pyplot)](https://matplotlib.org/3.5.0/api/_as_gen/matplotlib.pyplot.plot.html#matplotlib.pyplot.plot) para obtener más información y sus [ejemplos](http://matplotlib.org/gallery.html#statistics) para inspirarse.

Veamos un sencillo ejemplo de un programa que convierte de millas a kilómetros

```python
#ingreso las millas 
millas = [1, 2, 3]
conversor_a_km =1.60934 
km = []

km.append(conversor_a_km * millas[0])
km.append(conversor_a_km * millas[1])
km.append(conversor_a_km * millas[2])
print(km)
```

**Para pensar**: ¿Cómo podria mejorarse el programa anterior? ¿Existe alguna forma de evitar escribir 3 veces la misma línea de código? (*Pista*: se requiere aplicar una estructura de control)

```python
#importa la librería para graficar
import matplotlib.pyplot as plt 

plt.plot(millas, km) #plot() genera un gráfico de X contra Y
plt.show() #show() muestra el gráfico
```

Para entender las funciones, siempre es recomendable utilizar el comando de ayuda `?`, en donde veremos una descripción de lo que hace la función y de los parámetros que puede tomar.

```python
plt.plot?
```

Por ejemplo, en la función anterior, podemos agregar el parámetro '*' y se imprimirá un gráfico de puntos.

```python
plt.plot(millas, km, 'r+')
plt.show()
```

Podemos seguir indagando en otro tipo de gráficos.

```python
plt.figure(1) #permite ordenar la impresión de un conjunto de gráficos
plt.plot(millas, km, 'yp')
plt.grid() #muestra una grilla que orienta el gráfico

plt.figure(2) #de no incoporar esta línea, el primer gráfico no se plotea
plt.plot(millas, km, 'b<')
```

Podemos mejorar la impresión de los graficos agregando *capas* que especifiquen el nombre de los ejes, título, etc. El parámetro `label` me permite etiquetar a la variable


```python
import matplotlib.pyplot as plt
 
#cargo los valores de y1 e y2 que quiero graficar frente al eje x

y1 = [1, 3, 5, 3, 1, 3, 5, 3, 1] #variable 1
y2 = [2, 4, 6, 4, 2, 4, 6, 4, 2] #variable 2
x  = [1, 2, 3, 4, 5, 6, 7, 8, 9] #eje x

plt.plot(x, y1, ".-", color ='green', label="line B") #grafico la primera variable y la etiqueto
plt.plot(x, y2, 'r--', label="line D") #grafico la otra

plt.xlabel("x axis") #etiqueta del eje x
plt.ylabel("y axis") #etiqueta del eje y
plt.title("Line Graph Example") #titulo del grafico
plt.legend( loc = "upper left") #imprimo la legenda
```

```python
import numpy as np
import matplotlib.pyplot as plt

x_vec = np.linspace(-2*np.pi, 2*np.pi, 100)
sen_x = np.sin(x_vec)
cos_x = np.cos(x_vec)

plt.plot(x_vec, sen_x, color = "c", label = "sen(x)")
plt.plot(x_vec, cos_x, color = 'k', label = "cos(x)")
plt.title( "Función seno y coseno", fontsize = 12)
plt.legend(loc = "center")
 
```
