Tarea:

SE BUSCA CONOCER EL RENDIMIENTO DE LAS OBLIGACIONES NEGOCIABLES CLASE XIV DE VISTA ENERGY:

image.png

image.png

###### Fuente: https://vistaenergy.com/contenidos/1668628841.pdf

```python
#definición de variables
inversion = -1500  # deposito (ojo con el signo)
tasa_interes = .0625    # tasa de interes
cantidad_anios = 3      # cantidad de años
tasa_interes_total = (tasa_interes * cantidad_anios)
ganancia_total = inversion * -tasa_interes_total
interes_capital_total = float( (inversion + (inversion * tasa_interes_total)) * -1 )

print(f"""
      Suscribiendo al canje de la ON clase XIV, con Una inversión de {-1*inversion} USD, 
      una tasa anual del {tasa_interes}, en un plazo de {cantidad_anios} años;
      Se obtendría una gananacia del {ganancia_total} USD, 
      totalizando {interes_capital_total} USD.
      """)
```

```python
# definición de una función
def vf_on(inversion, tasa_interes, cantidad_anios):
  fv = npf.fv(pv=inversion, rate=tasa_interes, nper=cantidad_anios, pmt=0)
  return fv
```

```python
on_clase_xiv = vf_on(inversion, tasa_interes, cantidad_anios).round(2)
diff_ganancia = on_clase_xiv - interes_capital_total

print(f"""
      No obstante, se podría tener un rendimiento total de {round(on_clase_xiv,2)} USD,
      reinvirtiendo los intereses, es decir {round(diff_ganancia,2)} USD.      
      """)
```

#### Alternativa I

Voy a elegir *Sucesión de Fibonacci* ya que es la más famosa.
$ [0,1,1,2,3,5,8,13,21] $


```python
def FIB(n):
    h = 0
    g = 1
    for i in range(n):
       j = h + g
       h = g
       g = j
    return g

for i in range(0,11):
    if i == 0:
        print(0)
    elif (i > 0) & (i <= 2):
        print(1)
    else:
        print(FIB(i))
```

## Alternativa II

Le secuencia que he seleccionado es (1,1,4,7,19,40,...), la misma me ha resultado muy dificil de resolver en la practica, es por ello que, trataremos de abordar a los valores que la constituyen a traves de un programa que permita visualizar que valor corresponde a cada "n" o posicion en la secuencia. 

```python
# Para facilitar la interaccion, agregamos la funcion input que contemplara la posicion que deseamos hallar:
pos= input("Ingrese la posicion que desea conocer en la secuencia: ")
pos=int(pos)

# Posteriormente, planteamos la sucesion de la siguiente manera:
def secuencia(n):
    if n == 0:
        serie_secuencia = []
    elif n == 1:
        serie_secuencia = [1]
    elif n == 2:
        serie_secuencia = [1, 1]
    elif n > 2:
        serie_secuencia = [1, 1]
        i = 1

        while i < n - 1:
            serie_secuencia.append(serie_secuencia[i] + 3*serie_secuencia[i - 1])
            i += 1
    
    return serie_secuencia

# Hallamos el valor correspondiente a la posicion solicitada.
resultado = secuencia(pos+1)

print("El valor que corresponde a la posicion solicitada es:",resultado[pos])
```

```python
# Aqui podemos visualizar como esta compuesta la lista generada en el programa anterior. 
print(resultado)
```
